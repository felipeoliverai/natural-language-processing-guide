```
import nltk 
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('brown')

from nltk import FreqDist
from nltk.corpus import brown
```

```
# show categories 
print("\nCategorias no Corpus:")
print(brown.categories())
```

```
# list some categories 
word_categories = ['fiction', 'humor', 'romance']
```

```
# word that begin with "wh" sufix 
whwords = ['what', 'which', 'why', 'when', 'where', 'who']
```

```
# frequency distribution 
for i in range(0,len(word_categories)):
    genre = word_categories[i]
    print()
    print("\nAnalysis '"+ genre + "' and extract letters with sufix wh")
    genre_text = brown.words(categories=genre)
    fdist = FreqDist(genre_text)
    for wh in whwords:
        print(wh + ':', fdist[wh], end=' ')

print("\n")
```
